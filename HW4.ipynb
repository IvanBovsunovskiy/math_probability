__1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.__

Так как распределение равномерное, то среднее значение: $M = (800 + 200)/2 = 500$

Дисперсия равномерного распределения: $D(X) = \frac{(800 - 200)^2}{12} = 30000$


__2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
Если да, найдите ее.__

Из формулы для дисперсии равномерного распределения имеем:
$D(X) = \frac{(b - a)^2}{12} \rightarrow b = \sqrt{12\cdot D} + a = \sqrt{12\cdot 0.2} + 0.5 \approx 2.05$

__3. Непрерывная случайная величина X распределена нормально и задана плотностью распределения
$f(x) = (\frac{1}{4 \sqrt{2\cdot\pi}}) \cdot e^{-\frac{(x+2)^2}{32}}$.
Найдите:
а). M(X)
б). D(X)
в). std(X) (среднее квадратичное отклонение)__

Так как в общем виде нормальное распределение выражается: $$f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e ^ {- \frac{(x - a)^2}{2 \sigma^2}},$$ то
__а)__ $M(X) = -2$

__б)__ $D(X) = 32/2 = 16$

__в)__ $\sigma = \sqrt{D(X)} = 4$




__4. Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
а). больше 182 см
б). больше 190 см
в). от 166 см до 190 см
г). от 166 см до 182 см
д). от 158 см до 190 см
е). не выше 150 см или не ниже 190 см
ё). не выше 150 см или не ниже 198 см
ж). ниже 166 см.__

$F_{\mu,\sigma}(x) = F(\frac{x-\mu}{\sigma})$

__а)__ Вероятность того что взрослый человек, выбранный случайным образом, имеет рост больше 182см будет:
$P_{>182} = 1 - F(\frac{182-174}{8}) = 1-F(1) \approx 1 - 0.8413 = 0.1587$

__б)__ Вероятность того что взрослый человек, выбранный случайным образом, имеет рост больше 190см будет:
$P_{>190} = 1 - F(\frac{190-174}{8}) = 1-F(2) \approx 1 - 0.9772 = 0.0228$

__в)__ Вероятность того что взрослый человек, выбранный случайным образом, имеет рост от 166см до  190см будет:
$P_{>166, ~<190} = F(\frac{190-174}{8}) - F(\frac{166-174}{8}) = F(2) - F(-1) \approx 0.9772  - 0.1587 = 0.8185$

__г)__ Вероятность того что взрослый человек, выбранный случайным образом, имеет рост от 166см до  182см будет:
$P_{>166, ~<182} = F(\frac{182-174}{8}) - F(\frac{166-174}{8}) = F(1) - F(-1) \approx 0.8413  - 0.1587 = 0.6826$

__д)__ Вероятность того что взрослый человек, выбранный случайным образом, имеет рост от 158см до  190см будет:
$P_{>158, ~<190} = F(\frac{190-174}{8}) - F(\frac{158-174}{8}) = F(2) - F(-2) \approx 0.9772  - 0.0228 = 0.9544$

__е)__ Вероятность того что взрослый человек, выбранный случайным образом, имеет рост не более 150см или не менее 190см будет:
$P_{<150 ~\&~ >190} = F(\frac{150-174}{8}) + 1 - F(\frac{190-174}{8}) = F(-3) + 1 - F(2) \approx 0.0013  + 0.0228 = 0.0241$

__ё)__ Вероятность того что взрослый человек, выбранный случайным образом, имеет рост не более 150см или не менее 198см будет:
$P_{<150 ~\&~ >198} = F(\frac{150-174}{8}) + 1 - F(\frac{198-174}{8}) = F(-3) + 1 - F(3) \approx 0.0013  + 0.0013 = 0.0026$

__ж)__ Вероятность того что взрослый человек, выбранный случайным образом, имеет рост не более 166см будет:
$P_{<166} = F(\frac{166-174}{8}) = F(-1) \approx 0.1587$
