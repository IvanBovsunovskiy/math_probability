__1. Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks): zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110], ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]. Используя математические операции, посчитать коэффициенты линейной регрессии, приняв за X заработную плату (то есть, zp - признак), а за y - значения скорингового балла (то есть, ks - целевая переменная). Произвести расчет как с использованием intercept, так и без.__

```python
 zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110]).reshape((-1,1))
 ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
```

Расчет коэффициентов с использованием intercept

```python
linearmodel = LinearRegression(fit_intercept = True).fit(zp, ks)
"""fit_intercept — это логическое значение (по умолчанию True), которое определяет, следует ли вычислять точку пересечения b_0b 
0 (True) или считать ее равной нулю (False)"""
r_sq = linearmodel.score(zp, ks)
print('coefficient of determination:', r_sq)
print('intercept:', linearmodel.intercept_)
print('slope:', linearmodel.coef_[0])
```

Расчет коэффициентов с без использования intercept

```python
linearmodel = LinearRegression(fit_intercept = False).fit(zp, ks)
"""fit_intercept — это логическое значение (по умолчанию True), которое определяет, следует ли вычислять точку пересечения b_0b 
0 (True) или считать ее равной нулю (False)"""
r_sq = linearmodel.score(zp, ks)
print('coefficient of determination:', r_sq)
print('intercept:', linearmodel.intercept_)
print('slope:', linearmodel.coef_[0])
```

__2. Посчитать коэффициент линейной регрессии при заработной плате (zp), используя градиентный спуск (без intercept).__

```python
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
# количество шагов спуска
steps = 1000
# скорость обучения
learningrate = 1e-6
x = zp
y = ks
print(x,y)
# задаём начальные случайные значения коэффициентам линейной регрессии 
# b0 = np.random.randn(1) - c intercept
b0 = 0
b1 = np.random.randn(1)
print(f'Initial b1 - {b1[0]}')
for step in range(steps):
  
    # рассчитываем результирующий массив с текущими коэффициентами b0 и b1
    # на основе обучающей выборки 
    yhat = b0 + b1 * x
    
    # 1. определяем ошибку
    error = (y - yhat)
    
    # считаем градиенты (вспоминая формулу производной)
    # для коэффициента b0
    #b0_grad = -2 * error.mean()
    # для коэффициента b1
    b1_grad = -2 * (x * error).mean()
    
    # обновляем параметры, используя коэффициент скорости обучения
    #b0 = b0 - learningrate * b0_grad
    b1 = b1 - learningrate * b1_grad
    if step%50 == 0:
      print(b1)
      print(b1_grad)
print(f'Final b1 - {b1}')   
```

__3. В каких случаях для вычисления доверительных интервалов и проверки статистических гипотез используется таблица значений функции Лапласа, а в каких - таблица критических точек распределения Стьюдента?__

Таблица значений функции Лапласа - это вероятность того, что случайная величина примет значение, принадлежащее заданному интервалу. Используется при известных дисперсии и ожидаемом значении. Таблица критических точек распределения Стьюдента используется при неизвестной дисперсии генеральной совокупности и математическом ожидании, а известных дисперсии и мо выборки.

__*4. Произвести вычисления как в пункте 2, но с вычислением intercept. Учесть, что изменение коэффициентов должно производиться на каждом шаге одновременно (то есть изменение одного коэффициента не должно влиять на изменение другого во время одной итерации).__

```python
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
# количество шагов спуска
steps = 1000000
# скорость обучения
learningrate = 1e-5
x = zp
y = ks
print(x,y)
# задаём начальные случайные значения коэффициентам линейной регрессии 
b0 = np.random.randn(1)
b1 = np.random.randn(1)

"""b0 = 444
b1 = 2.6"""
print(f'Initial b0 - {b0} \n Initial b1 - {b1}')
for step in range(steps):
  
    # рассчитываем результирующий массив с текущими коэффициентами b0 и b1
    # на основе обучающей выборки 
    yhat = b0 + b1 * x
    
    # 1. определяем ошибку
    error = (y - yhat)
    
    # считаем градиенты (вспоминая формулу производной)
    # для коэффициента b0
    b0_grad = -2 * error.mean()
    # для коэффициента b1
    b1_grad = -2 * (x * error).mean()
    
    # обновляем параметры, используя коэффициент скорости обучения
    b0 = b0 - learningrate * b0_grad
    b1 = b1 - learningrate * b1_grad
    if step%50000 == 0:
      print(f'Step - {step}  b0 - {b0}  b1 - {b1}  b0_grad - {b0_grad}   b1_grad - {b1_grad}')
      
print(f'Final b0 - {b0} \n Final b1 - {b1}')
```
