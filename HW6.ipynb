__1. Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks):
zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110],
ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832].
Найдите ковариацию этих двух величин с помощью элементарных действий, а затем с помощью функции cov из numpy
Полученные значения должны быть равны.
Найдите коэффициент корреляции Пирсона с помощью ковариации и среднеквадратичных отклонений двух признаков,
а затем с использованием функций из библиотек numpy и pandas.__

Так как $cov_{XY} = M(XY) - M(X)\cdot M(Y)$, то

```python
zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110]
ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]
```

```python
Mzp = np.mean(zp)
Mks = np.mean(ks)
zpks = [x*y for x,y in zip(zp,ks)]
Mzpks = np.mean(zpks)
cov = Mzpks - Mzp*Mks
print(cov)
```

Или с помощью функции numpy

```python
print(np.cov(zp,y = ks, ddof = 0))
```

Так как коэффициент корелляции Пирсона $r_{XY} = \frac{cov_{XY}}{\sigma_X\cdot\sigma_Y}$, то

```python
rxy = cov/(np.std(zp)*np.std(ks))
print(f'Коэффициент корелляции Пирсона: {rxy}')
```

```python
print(f'Коэффициент корелляции Пирсона из библиотеки numpy: {np.corrcoef(zp, y=ks)[1][0]}')
```

__2. Измерены значения IQ выборки студентов,
обучающихся в местных технических вузах:
131, 125, 115, 122, 131, 115, 107, 99, 125, 111.
Известно, что в генеральной совокупности IQ распределен нормально.
Найдите доверительный интервал для математического ожидания с надежностью 0.95.__

```python
iq = [131, 125, 115, 122, 131, 115, 107, 99, 125, 111]
```

Так как не известны значения ско и мат ожидания то используем t-распределение. Тогда границы интервала будут для t-распределения с 9 степенями свободы и уровнем значимости для двухстороннего распределения 0.025:
$$T_{1,2} = \overline{X} \pm \frac{s_0}{\sqrt{n}} \cdot t_{\alpha/2} = 118.1 \pm \frac{10.55}{\sqrt{10}} \cdot 2.2622 \Rightarrow (110.56, 125.64)$$


```python
Mx = np.mean(iq)
sigma = np.std(iq, ddof=1)
print(f'Математическое ожидание - {Mx},\nСреднее квадратичное отклонение - {sigma}, \nГраницы интервала - ({Mx - sigma/10**0.5*2.2622},{Mx + sigma/10**0.5*2.2622})')
```
